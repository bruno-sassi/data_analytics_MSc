```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(0, 50, 400)

# Define constraint functions
y1 = (120 - 5*x) / 3        # from 5x + 3y <= 120
y2 = (100 - 3*x) / 4        # from 3x + 4y <= 100
y3 = np.full_like(x, 25)    # y <= 25
x_min = 10                  # x >= 10

# Plot each constraint line
plt.plot(x, y1, label=r'$5x+3y \leq 120$')
plt.plot(x, y2, label=r'$3x+4y \leq 100$')
plt.axhline(25, color='purple', linestyle='--', label=r'$y \leq 25$')
plt.axvline(10, color='green', linestyle='--', label=r'$x \geq 10$')

# Fill feasible region
y_upper = np.minimum.reduce([y1, y2, y3])  # take the min of all upper bounds
plt.fill_between(x, 0, y_upper, where=(x>=x_min), color='lightgrey', alpha=0.5)

# Labels and legend
plt.xlim(0, 50)
plt.ylim(0, 50)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Feasible Region for Production")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x range
x = np.linspace(0, 12, 400)

# Constraint lines
y1 = (40 - 4*x) / 2      # y <= 20 - 2x
y2 = (50 - 3*x) / 5      # y <= 10 - 0.6x
y3 = np.full_like(x, 12) # y <= 12

# Vertices of feasible polygon
v1 = (0.0, 0.0)
v2 = (10.0, 0.0)
x12 = 10 / 1.4
y12 = 20 - 2*x12
v3 = (x12, y12)
v4 = (0.0, 10.0)

poly_x = [v1[0], v2[0], v3[0], v4[0]]
poly_y = [v1[1], v2[1], v3[1], v4[1]]

# Feasible point
pt = (8, 4)

plt.figure(figsize=(7, 6))

# Plot constraint boundary lines
plt.plot(x, y1, label="4x + 2y = 40")
plt.plot(x, y2, label="3x + 5y = 50")
plt.plot(x, y3, label="y = 12")
plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)

# Shade feasible region
plt.fill(poly_x, poly_y, alpha=0.3, label="Feasible region")

# Mark vertices
for vx, vy in zip(poly_x, poly_y):
    plt.scatter([vx], [vy])
    plt.text(vx + 0.1, vy + 0.1, f"({vx:.2f}, {vy:.2f})")

# Highlight the point (8,4)
plt.scatter(*pt, color="red", s=80, zorder=0, label="(8,4)")
plt.text(pt[0] + 0.2, pt[1] - 0.3, "(8,4)", color="red")

plt.xlim(0, 12)
plt.ylim(0, 12.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Feasible Region with Point (8,4)")
plt.legend(loc="upper right")
plt.grid(True)

plt.show()
```
